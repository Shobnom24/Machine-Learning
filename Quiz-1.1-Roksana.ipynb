```python
# Q6: Replace the value 2 from column "sex" with 0. Initially 1 stands for male and 2 for female.

replaced_value = df["sex"].replace(2, 0)
print(replaced_value)
```

```python
# Q7: Find all the males with whose age is between 30 to 40

filtered_value = df[(df.sex == 1) & (df["age"].between(30, 40))]
print(filtered_value)
```

```python
# Q8: Remove the columns "weight2" and "wtyrago"

removed_columns = df.drop(columns=['weight2', 'wtyrago'])
print(removed_columns)
```
